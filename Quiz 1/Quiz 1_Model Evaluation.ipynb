## Metrics to Evaluate

I calculate accuracy, recall, precision and f1 score to evaluate the training model.

````python
TP, FP, TN, FN = 1173, 33, 2909, 21

# Accuracy
accuracy = (TP + TN) / (TP + FP + FN + TN)

# Recall(Sensitivity)
recall = TP / (TP + FN)

# Precision
precision = TP / (TP + FP)

# F1 Score
f1_score = 2 * (recall * precision) / (recall + precision)

accuracy, recall, precision, f1_score
````

## Results and Evaluation

The model has 98% accuracy. Accuracy is a good metric to evaluate the model when the dataset is symmetric which means values of false positive and false negatives are almost same.

The ratio of the recall reflects the percentage correctly found. I think we need to use recall to evaluate our model in the spam email problem because we focus on how correctly we predict spam emails to the all positive cases.
